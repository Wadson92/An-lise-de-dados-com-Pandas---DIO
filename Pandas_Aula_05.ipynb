# Trabalhando com Planilhas do Excel

```python
#Importando a biblioteca Pandas
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("/content/drive/MyDrive/Datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Datasets/Salvador.xlsx")
```

```python
df1.head()
```

```python
#Juntando todos os aquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
#Para pegar uma amostra do nosso conjunto de dados
df.sample(5)
```

```python
#Alterar o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

### Tratando valores faltantes

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
df
```

```python
df["Vendas"].isnull().sum()
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagar linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

### Criando colunas novas

```python
#Criando a coluna da receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Se eu não tivesso a coluna Qtde, eu poderia descobrir dividindo vendas por receitas
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df['Receita'].max()
```

```python
#Quando quero retornar todas as referencias com base nas maiores receitas
df.nlargest(3, "Receita")
```

```python
#Quando quero retornar todas as referencias com base nas menores receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidades
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```
