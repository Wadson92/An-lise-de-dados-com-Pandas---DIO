```python
#Criando nosso DataFrame
df = pd.read_excel("/content/drive/MyDrive/Datasets/AdventureWorks.xlsx")
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando a coluna de custo
```

```python
df.head()
```

```python
#Qual o custo total?
round(df["Custo"].sum(), 2)
```

```python
#Qual o lucro total? Agora que temos receita e custo e o total.
#Vamos criar uma coluna de lucro que será Receita - Custo
df["Lucro"] = df['Valor Venda'] - df['Custo']
```

```python
df.head(1)
```

```python
#Total de lucro
round(df['Lucro'].sum(), 2)
```

```python
#Criando uma coluna com o total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

###### Média do tempo de envio para cada marca, e para isso precisamos transformar a coluna Tempo_envio em numérica

```python
#Extraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
df.dtypes
```

```python
#Média de tempo de envio por Marca
df.groupby('Marca')['Tempo_envio'].mean()
```

##### Missing Values

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

###### Se quisermos saber o Lucro por ano e Marca?

```python
#Vamos agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
#Formando os valores mostrados na tela
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
#Analisando as vendas do ano de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Analisando lucro por mês de 2009
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title="Lucro x Mês", marker="v")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Lucro por marca
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal'); #Ele altera o eixo x para a horizontal
```

```python
#Lucro por classe
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro x Classe')
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
